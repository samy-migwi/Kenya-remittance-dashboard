```python

import pandas as pd
import plotly.express as px

# Sample DataFrame structure (replace this with your actual DataFrame)
# df = pd.read_csv('your_data.csv')  # Load your actual DataFrame here

# Assuming you have columns from 'Jan_20' to 'Aug_24'
columns_of_interest = [
    'Region/Country',
    'Jan_20', 'Feb_20', 'Mar_20', 'Apr_20', 'May_20', 'Jun_20', 'Jul_20', 'Aug_20',
    'Jan_21', 'Feb_21', 'Mar_21', 'Apr_21', 'May_21', 'Jun_21', 'Jul_21', 'Aug_21',
    'Jan_22', 'Feb_22', 'Mar_22', 'Apr_22', 'May_22', 'Jun_22', 'Jul_22', 'Aug_22',
    'Jan_23', 'Feb_23', 'Mar_23', 'Apr_23', 'May_23', 'Jun_23', 'Jul_23', 'Aug_23',
    'Jan_24', 'Feb_24', 'Mar_24', 'Apr_24', 'May_24', 'Jun_24', 'Jul_24', 'Aug_24'
]

# Filter the DataFrame to keep only the relevant columns
df_filtered = df[columns_of_interest]

# Melt the DataFrame to long format for easier manipulation
df_long = df_filtered.melt(id_vars=['Region/Country'], var_name='Month', value_name='Value')

# Extract the year and month for better plotting
df_long['Year'] = df_long['Month'].str.split('_').str[1]
df_long['Month'] = df_long['Month'].str.split('_').str[0]

# Convert 'Value' to numeric, forcing errors to NaN
df_long['Value'] = pd.to_numeric(df_long['Value'], errors='coerce')

# Filter out rows where 'Value' is NaN or zero
df_long = df_long[(df_long['Value'].notna()) & (df_long['Value'] != 0)]

# Create Sunburst charts for each year
years = df_long['Year'].unique()

for year in years:
    year_data = df_long[df_long['Year'] == year]
    
    # Check if there is data for the year
    if year_data.empty:
        print(f"No data available for year {year}. Skipping this year.")
        continue

    # Create the Sunburst chart
    fig = px.sunburst(
        year_data,
        path=['Year', 'Month', 'Region/Country'],  # Define the hierarchy
        values='Value',
        height=1500,
        width=1200,
        title=f'Sunburst Chart for Year {year}',
        color='Value',  # Color based on the value
        color_continuous_scale='RdBu'
    )
    
    # Show the figure
    fig.show()
```
