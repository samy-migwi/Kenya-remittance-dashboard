```python
#clean 1
# removing first unnamed column
dt=dt.drop(columns="Unnamed: 0")
dt.head()
```

```python
#clean 2
# Drop columns where all values are NaN
dt = dt.dropna(axis=1, how='all')

#clean 3
# Rename the columns based on expected structure
dt.columns = ['Region/Country', 'Year', 'Month'] + [f'Month_{i}' for i in range(1, len(dt.columns) - 3 + 1)]

# Fill in the 'Year' column
dt['Year'] = dt['Year'].ffill()

# Drop rows with unnecessary information
dt = dt.dropna(subset=['Region/Country'])

dt.head()
```

```python
dt.head(10)
```

```python
#this will drop the months of the year 2019 , we wont proceed with it since it is not full . Data from jan is missing
dt = dt.drop(dt.columns[1:12], axis=1)
dt.head()
```

```python
# we start by renaming the columns 
#define the starting month and year
start_month = 1  
start_year = 20  

#  N.B in the cell above we have droped data for 2019 so that we can use the data from 2020 since jan 2019 data is not available
# generate the month and year labels
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
new_columns = ["Region/Country"] + [f"{months[(start_month - 1 + i) % 12]}_{start_year + (start_month - 1 + i) // 12}" for i in range(len(dt.columns) - 1)]

# Rename the columns
dt.columns = new_columns


dt.head()
```

```python
# Drop rows where the "Region/Country" column matches any entry in region_list
dt = dt[dt['Region/Country'].str.contains('Total', case=False, na=False)]


# Reset the index after dropping rows
dt.reset_index(drop=True, inplace=True)

# Display the first few rows to verify the changes
dt.head()
```
