```python

# Droping NaN columns
df_cleaned = df_import.dropna(axis=1, how='all')

# Set the second row (index 1) as the header
df_cleaned.columns = df_cleaned.iloc[1]

# Droping the rows used for headers and irrelevant NaN rows
df_cleaned = df_cleaned.drop([0, 1]).reset_index(drop=True)

# Step 4: Manually renaming the columns for the months with years
month_columns = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
year_suffix = ['_20', '_21', '_22', '_23', '_24']  # Define year suffixes

# new column names from Jan_20 to Aug_24
new_column_names = ['Region/Country']  # Keep the first column name as is

# Generate month names with year suffixes
for suffix in year_suffix:
    if suffix == '_24':
        new_column_names += [f'{month}{suffix}' for month in month_columns[:8]] 
    else:
        new_column_names += [f'{month}{suffix}' for month in month_columns] 

# Assign the new column names
df_cleaned.columns = new_column_names

df=df_cleaned

```

```python
import plotly.express as px

# Sample data structure for your DataFrame (df)
fig = px.line(df, x=df.columns[1:], y=df.iloc[:, 1:], 
              labels={"variable": "Month", "value": "Remittances (USD)", "Region/Country": "Country"}, 
              title="Monthly Remittances by Country (Jan 2020 - Aug 2024)",
              facet_row="Region/Country")

fig.update_layout(height=600, width=1000)
fig.show()
```

```python
# Create heatmap
fig = px.imshow(df.pivot('Region/Country', 'Month', 'Remittances'),
                labels={"x": "Month", "y": "Country", "color": "Remittances (USD)"},
                title="Heatmap of Monthly Remittances by Country")

fig.update_layout(height=600, width=1000)
fig.show()
```
